2.How does broadcasting work in NumPy?
- Broadcasting in NumPy allows arithmetic operations on arrays of different shapes without explicitly reshaping them. Instead of creating duplicate data, NumPy automatically expands smaller arrays to match larger ones, making computations more efficient.

Broadcasting Rules
- When performing operations on arrays of different shapes, NumPy follows these rules:

- Match Dimensions from Right to Left - Shapes are compared from the last (rightmost) dimension.

- Dimensions Must Be Equal or One of Them Must Be 1 - If two dimensions differ, one must be 1 for broadcasting to work.

6. What are the differences between NumPy arrays and Python lists?
- 1. Performance
- NumPy Arrays: Faster because they use optimized C-based implementations and store data in a contiguous block of memory.
- Python Lists: Slower since they store references to objects, causing overhead in type checking and memory allocation.
2. Memory Efficiency
- NumPy Arrays: More memory-efficient as they store elements of the same data type in a continuous block.
- Python Lists: Less efficient because they store pointers to objects, increasing memory consumption.
3. Data Type Consistency
- NumPy Arrays: Homogeneous (all elements must be of the same data type).
- Python Lists: Heterogeneous (can store different data types within the same list.
